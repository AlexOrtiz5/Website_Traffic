# Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px  # For interactive plots
```

```python
# Load your cleaned data
df = pd.read_csv("./Data/clean_website_wata.csv")
df
```

# 1. Pair Plots

```python
sns.pairplot(df[['Page Views', 'Session Duration', 'Bounce Rate', 'Conversion Rate']])
plt.suptitle('Pair Plots of Numerical Variables', y=1.02)
plt.show()
```

# 2. Heatmaps (Correlation Matrix)

```python
correlation_matrix = df[['Page Views', 'Session Duration', 'Bounce Rate', 'Conversion Rate', 'Previous Visits', 'Time on Page']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

# 3. Interactive Plots (Plotly Express - Scatter Plot)

```python
fig = px.scatter(df, x='Session Duration', y='Page Views', color='Traffic Source',
                 title='Interactive Scatter Plot: Session Duration vs. Page Views')
fig.show()
```

# 4. Interactive Plots (Plotly Express - Histogram)

```python
fig = px.histogram(df, x='Bounce Rate', color='Traffic Source',
                   title='Interactive Histogram: Bounce Rate Distribution by Traffic Source')
fig.show()
```

# 5. Interactive Plots (Plotly Express - Box Plot)

```python
fig = px.box(df, x='Traffic Source', y='Conversion Rate',
             title='Interactive Box Plot: Conversion Rate by Traffic Source')
fig.show()
```

# 6. Funnel Chart (Example - if you had funnel steps, you'd adapt this)

```python
# Example of making a fake funnel dataset. You would adapt this to your own funnel data.
funnel_data = pd.DataFrame({
    'Step': ['Landing Page', 'Product View', 'Add to Cart', 'Checkout', 'Purchase'],
    'Users': [1000, 800, 500, 300, 200]
})

fig = px.funnel(funnel_data, x='Users', y='Step', title='Example Funnel Chart')
fig.show()
```

# 7. Time series interactive plot (if you have the time data).

```python
# Example of making a fake time series dataset. You would adapt this to your own time series data.
time_data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=365),
    'Page Views': [i + 50 * np.sin(i / 10) + np.random.randint(-50, 50) for i in range(365)]
})

fig = px.line(time_data, x='Date', y='Page Views', title='Time Series: Page Views Over Time')
fig.show()
```
