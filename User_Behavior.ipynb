# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
df = pd.read_csv("./Data/clean_website_wata.csv") 
df
```

# Bounce rate analysis

```python
print(f"Overall Bounce Rate: {df['Bounce Rate'].mean() * 100:.2f}%")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Bounce Rate'], kde=True)
plt.title('Distribution of Bounce Rates')
plt.xlabel('Bounce Rate')
plt.ylabel('Frequency')
plt.show()

# Save the figure
plt.savefig("./Behavior_Graphs/Distribution_Rates.png")  # Saves as PNG (default)
```

# Time on site analysis

```python
print(f"Average Time on Site: {df['Session Duration'].mean():.2f} seconds")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Session Duration'], kde=True)
plt.title('Distribution of Session Durations')
plt.xlabel('Session Duration (seconds)')
plt.ylabel('Frequency')
plt.show()

# Save the figure
plt.savefig("./Behavior_Graphs/Distribution_Durations.png")  # Saves as PNG (default)
```

# Pages per session analysis

```python
print(f"Average Pages per Session: {df['Page Views'].mean():.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Page Views'], kde=True)
plt.title('Distribution of Page Views')
plt.xlabel('Page Views')
plt.ylabel('Frequency')
plt.show()

# Save the figure
plt.savefig("./Behavior_Graphs/Distribution_Views.png")  # Saves as PNG (default)
```

# Conversion rate analysis

```python
print(f"Overall Conversion Rate: {df['Conversion Rate'].mean() * 100:.2f}%")
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Conversion Rate'], kde=True)
plt.title("Distribution of Conversion Rates")
plt.xlabel("Conversion Rate")
plt.ylabel("Frequency")
plt.show()

# Save the figure
plt.savefig("./Behavior_Graphs/Distribution_Conversion.png")  # Saves as PNG (default)
```

```python
conversion_previous_visits = df.groupby('Previous Visits')['Conversion Rate'].mean()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=conversion_previous_visits.index, y = conversion_previous_visits.values)
plt.title("Conversion rate based on previous visits")
plt.xlabel("Previous Visits")
plt.ylabel("Conversion Rate")
plt.show()

# Save the figure
plt.savefig("./Behavior_Graphs/Conversion_Rates.png")  # Saves as PNG (default)
```
