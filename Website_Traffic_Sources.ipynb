# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
df = pd.read_csv("./Data/clean_website_wata.csv") 
df
```

# Traffic source distribution

```python
traffic_source_counts = df['Traffic Source'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=traffic_source_counts.index, y=traffic_source_counts.values)
plt.title('Traffic Source Distribution')
plt.xlabel('Traffic Source')
plt.ylabel('Number of Sessions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Traffic source percentage distribution

```python
plt.figure(figsize=(8,8))
plt.pie(traffic_source_counts, labels = traffic_source_counts.index, autopct='%1.1f%%')
plt.title("Percentage of traffic per source")
plt.show()
```

# Conversion rates per traffic source

```python
conversion_rates = df.groupby('Traffic Source')['Conversion Rate'].mean().sort_values(ascending=False)
print("Conversion Rates by Traffic Source:\n", conversion_rates)
```
